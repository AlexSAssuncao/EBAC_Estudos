````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
df = pd.read_csv('previsao_de_renda.csv')
````

````python
plt.figure(figsize=(10, 6))
sns.pointplot(x='posse_de_veiculo',y='renda', data=df, errorbar=('ci', 95),label='posse_de_veiculo')
sns.pointplot(x='posse_de_imovel',y='renda', data=df, errorbar=('ci', 95),label='posse_de_imovel')

plt.show()
````
