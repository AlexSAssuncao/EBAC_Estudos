# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

```python
#1.a) Calculando média e diferença
df['média'] = df.mean(axis=1)
df['diferença'] = df.p2-df.p1
df.head()
```

```python
#1.b) correlação
correlacao = df['p1'].corr(df['p2'])# Calculando a correlação da matriz com colunas p1 e p2
print(f"A correlação entre p1 e p2 é: {correlacao}")
plt.scatter(x = df.p1, y=df.p2);# Plotando o gráfico de dispersão
```

```python
#1.c) correlação
correlacao = df['média'].corr(df['diferença'])# Calculando a correlação da matriz com colunas média e diferença
print(f"A correlação entre média e diferença é: {correlacao}")
plt.scatter(x = df['média'], y=df['diferença']);# Plotando o gráfico de dispersão
```

```python
#1.d) correlação observadas
plt.scatter(x = df.p1, y=df.p2, label='p1 e p2');
plt.scatter(x = df['média'], y=df['diferença'], label='média e diferença');
plt.legend();
print('A correlação observada entre p1 e p2 é maior que a correlação observada entre média e diferença')
```

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
df['soma'] = df.p1+df.p2
df.head()
```

```python
#correlação entre some e diferença
correlacao = df['soma'].corr(df['diferença'])# Calculando a correlação da matriz com colunas soma e diferença
print(f"A correlação entre soma e diferença é: {correlacao}, o que é o mesmo valor q a média e diferença")
plt.scatter(x = df['soma'], y=df['diferença']);# Plotando o gráfico de dispersão
```

```python
plt.scatter(x = df.p1, y=df.p2, label='p1 e p2');
plt.scatter(x = df['soma'], y=df['diferença'], label='soma e diferença');
plt.legend();
print('A disperção observada entre soma e diferença é o dobro no eixo X que a disperção observada entre p1 e p2')
```
