# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_raw[['munResNome','IDADEMAE','IDADEPAI']].groupby('munResNome',as_index=False).mean()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_aniv = sinasc_raw[['SEXO','PESO','ESCMAE']][sinasc_raw.DTNASC == '2019-10-29']
sinasc_aniv.groupby(['SEXO','ESCMAE']).mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
#contagem de nascidos municípios
sinasc_raw.munResNome.value_counts()
```

```python
print('Município com menos nascimentos é: ')
sinasc_raw.munResNome.value_counts().index[-2]
```

```python
#idade media, maxima, minima das maes e pais desse município
sinasc_item3 = sinasc_raw[sinasc_raw.munResNome == 'Castanheiras']
sinasc_item3[['munResNome','IDADEMAE','IDADEPAI']].groupby('munResNome').agg({'IDADEMAE':['min','mean','max'],'IDADEPAI':['min','mean','max']})
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
#FILTRANDO nascidos no mês de Março
sinasc_marco = sinasc_raw[sinasc_raw.DTNASC.map(lambda x:'-03-' in x )]
sinasc_marco.DTNASC.head()
```

```python
#Município de maior incidência de nascidos
sinasc_marco.munResNome.value_counts().index[0]
```

```python
## Min, Média e Max quantidade filhos vivos
sinasc_marco[sinasc_marco.munResNome == 'Porto Velho'][['munResNome','QTDFILVIVO']].groupby('munResNome', as_index=False).agg({'QTDFILVIVO':['min','mean','max']})
```

```python
## ## Min, Média e Max idade pais
sinasc_marco[sinasc_marco.munResNome == 'Porto Velho'][['munResNome','IDADEMAE','IDADEPAI']].groupby('munResNome', as_index=False).agg({'IDADEMAE':['min','mean','max'],'IDADEPAI':['min','mean','max']})
```
